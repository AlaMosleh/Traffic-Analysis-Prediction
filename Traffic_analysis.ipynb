9) Busiest Hours of the Day:

```python
from dateutil import parser
# Extract the hour from the Time column using dateutil.parser
df['Hour'] = df['Time'].apply(lambda x: parser.parse(x).hour)

# Group by the hour and calculate the total vehicle count for each hour
total_by_hour = df.groupby('Hour')['Total'].sum().reset_index()

# Plotting the total vehicle count by hour using a bar chart
plt.figure(figsize=(9, 6))
sns.barplot(x='Hour', y='Total', data=total_by_hour, palette='viridis')
plt.title('Total Vehicle Count by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Vehicle Count')
plt.show()

# Plotting the total vehicle count by hour and day using a heatmap
heatmap_data = df.pivot_table(index='Hour', columns='Day of the week', values='Total', aggfunc='sum')
plt.figure(figsize=(9, 6))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='g')
plt.title('Heatmap of Total Vehicle Count by Hour and Day')
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the Day')
plt.show()
```

How does the variance in vehicle counts compare across vehicle types

```python
vehicle_variances = df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].var()

plt.figure(figsize=(9, 6))
vehicle_variances.plot(kind='bar',color='navy')
plt.title('Variance in Vehicle Counts')
plt.xlabel('Vehicle Type')
plt.ylabel('Variance')
plt.xticks(rotation=0)
plt.show()
```

# Insights from the chart:

1. The CarCount variance is significantly higher than all other vehicle types.This suggests that the number of cars fluctuates widely over time.

2. BikeCount, BusCount, and TruckCount show much lower variance compared to cars.This indicates that their numbers remain relatively stable over time.

How does the average vehicle count for each type change over time

```python
import plotly.express as px
# plotly creates an interactive chart
# Reshape data: Convert wide format into long format making it easier to plot different vehicle types in one graph.

df_melted = df.melt(id_vars=['Time'], value_vars=['CarCount', 'BikeCount', 'BusCount', 'TruckCount'],
                     var_name='Vehicle Type', value_name='Count')

# Compute the average count per time for each vehicle type
df_grouped = df_melted.groupby(['Time', 'Vehicle Type'])['Count'].mean().reset_index()

# Create an interactive line chart

fig = px.line(df_grouped, x='Time', y='Count', color='Vehicle Type',
              title='Average Vehicle Counts Over Time',
              labels={'Count': 'Average Count', 'Time': 'Time'},
              markers=True)
fig.update_layout(width=1000,height=600)

fig.show()
```

#Insights from the chart

1. Car Count (green line)

  The car count fluctuates significantly over time.
  There is a noticeable increase in the count after a specific time (around 4:00 AM), indicating a surge in vehicle movement.
  After peaking, there is a slight decline towards the end.

2. Bike Count (Blue Line)

  The bike count remains relatively low throughout the timeline.
  there is a small fluctuations but stays within a narrow range.

3. Bus Count (Red Line)

  The bus count follows a pattern of low and moderate peaks, possibly representing scheduled bus routes.

4. Truck Count (Purple Line)

  The truck count remains relatively stable but exhibits small fluctuations.
  At certain points, there is a slight drop in truck numbers, which could indicate specific time restrictions for trucks.
